```python
Q1 = data['IC50'].quantile(0.25)
Q3 = data['IC50'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_array = np.where(data['IC50']>=upper)[0]  #нормализация данных
lower_array = np.where(data['IC50']<=lower)[0]
data.drop(index=upper_array, inplace=True)
data.drop(index=lower_array, inplace=True)
```
