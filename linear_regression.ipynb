```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("1400.xlsx - Structures-ADMET.csv")
```

```python
df = df.drop(columns=['Pictures','molecule', 'Unnamed: 55', 'Unnamed: 56','SI', 'CC50-MDCK, mmg/ml'])
```

```python
df = df.drop_duplicates(subset = 'SMILES')
```

```python
df = df.dropna()
```

```python
df = df.drop(columns=['SMILES'])
```

```python
df = df.replace(',', '.', regex=True)
```

```python
df
```

```python
IC50 = df[['y']]
X = df.drop(columns = ['y'])
```

```python
y = IC50
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Создайте экземпляр модели
model = LinearRegression()

# Обучите модель на обучающих данных
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Средняя квадратичная ошибка (MSE): {mse}")
print(f"Коэффициент детерминации (R-squared): {r2}")
```
